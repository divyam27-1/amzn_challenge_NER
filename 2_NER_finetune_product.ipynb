```python
cv_data = json.load(open('merged_hwd_annotations.json','r'))
annotations = cv_data["annotations"]
len(annotations)
```

```python
!python -m spacy init fill-config "./base_config.cfg" "./config.cfg"
```

```python
nlp = spacy.load('en_core_web_lg')
doc_bin = DocBin()
for training_example in tqdm(annotations):
    text = training_example[0]
    labels = training_example[1]["entities"]
    doc = nlp.make_doc(text)
    ents = []
    for start, end, label in labels:
        span = doc.char_span(start, end, label=label, alignment_mode="contract")
        if span is None:
            print("Skipping entity")
        else:
            ents.append(span)
    filtered_ents = filter_spans(ents)
    doc.ents = filtered_ents
    doc_bin.add(doc)


doc_bin.to_disk("train.spacy")
```

```python
import csv
import re
from concurrent.futures import ProcessPoolExecutor, as_completed
import math
```

```python
def apply_ner(text):
    #input cleaning
    text = str(text).lower()
    text = re.sub(r'[^a-z0-9.,\"\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()

    doc = nlp_ner(text)
    entities = [(ent.label_, ent.text) for ent in doc.ents]

    return tuple(entities) if entities else None

df = pd.read_csv("final_testdata.csv")

def process_and_save_chunk(chunk_df, chunk_number):
    chunk_df['ner_results'] = chunk_df['text'].apply(apply_ner)
    chunk_df.to_csv(f"chunk_{chunk_number}.csv", index=False)

chunk_size = math.ceil(len(df) / 30)

start_chunk = 1
end_chunk = 30

def process_chunk(start_index):
    end_index = min(start_index + chunk_size, len(df))
    chunk_df = df.iloc[start_index:end_index]
    chunk_number = (start_index // chunk_size) + 1
    if start_chunk <= chunk_number <= end_chunk:
        process_and_save_chunk(chunk_df, chunk_number)

with ProcessPoolExecutor() as executor:
    futures = [executor.submit(process_chunk, i) for i in range(0, len(df), chunk_size)]
    for future in as_completed(futures):
        # Optionally handle exceptions or get results here
        pass
```

```python
with open("data_1k.csv") as infile, open('ner_data_1k.csv', 'w', newline='') as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)

    header = next(reader)
    header.extend(['label', 'entity_text', 'start_char', 'end_char'])
    writer.writerow(header)

    for row in reader:

        text = row[4]
        text = str(text)
        text = text.lower()
        text = re.sub(r'[^a-z0-9.,\"\s]', ' ', text)
        text = re.sub(r'\s+', ' ', text).strip()

        doc = nlp_ner(text)

        for ent in doc.ents:
            print([ent.label_, ent.text, ent.start_char, ent.end_char])
            row.extend([ent.label_, ent.text, ent.start_char, ent.end_char])
            writer.writerow(row)
```
