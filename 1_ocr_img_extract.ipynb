```python
from paddleocr import PaddleOCR
import paddle
import requests
from PIL import Image
from io import BytesIO
from matplotlib import pyplot as plt

ocr = PaddleOCR(use_angle_cls=True, lang="en")
```

```python
import pandas as pd
data = pd.read_csv("train.csv")
```

```python
import numpy as np
```

```python
results = []
for i in range(2500, 3000):
    link = data['image_link'][i]
    image_url = str(link)
    group_id = data['group_id'][i]
    entity_name = data['entity_name'][i]
    entity_value = data['entity_value'][i]
    response = requests.get(image_url)
    if response.status_code == 200:


        image_bytes = BytesIO(response.content)
        img = Image.open(image_bytes)
        try:
            img = img.convert("L")
            img = np.array(img)
            result = ocr.ocr(img, cls=True)
        except:
            results.append({'index':i,'text': "",'group_id':group_id,'entity_name':entity_name,'entity_value':entity_value})
            continue
        if result and result[0]:
            sentence = []
            # Collect extracted text and confidence scores
            for line in result[0]:
                text = line[1][0]
                confidence = line[1][1]
                sentence.append(text)  # Append each word to the sentence list
        full_sentence = ' '.join(sentence)

            # Append the sentence and confidence to the results list
        results.append({'index':i,'text': full_sentence,'group_id':group_id,'entity_name':entity_name,'entity_value':entity_value})
        print(i)
        print(full_sentence)
```

```python
results_df = pd.DataFrame(results)

merged_df = pd.merge(data, results_df, left_index=True, right_on='index', how='left')

merged_df = merged_df.drop(columns=['index'])

merged_df.to_csv('combined_data_2k5_3k.csv', index=False)
```
